```python
groq_api_key=os.getenv("GROQ_API_KEY")
groq_api_key
```

```python
from langchain_groq import ChatGroq
model=ChatGroq(model="Gemma2-9b-it",groq_api_key=groq_api_key)
model
```

```python
from langchain_core.messages import HumanMessage
model.invoke([HumanMessage(content="Hi,My name is Niraj and I am a chief AI engineer")])
```

```python
from langchain_core.messages import AIMessage
model.invoke(
    [
        HumanMessage(content="Hi, My name is Niraj and I am Chief AI engineer"),
        AIMessage(content="Hi Niraj, it's nice to meet you!  That's a fascinating role.  \n\nWhat kind of AI projects are you working on?  I'm always interested to hear about the innovative things people are doing with artificial"),
        HumanMessage(content="Hey Whats my name and what do I do?")
    ]
)
```

```python
from langchain_community.chat_message_histories import ChatMessageHistory
from langchain_core.chat_history import BaseChatMessageHistory
from langchain_core.runnables.history import RunnableWithMessageHistory
store={}
def get_session_history(session_id: str)-> BaseChatMessageHistory:
    if session_id not in store:
        store[session_id] =ChatMessageHistory()
    return store[session_id]
with_message_history= RunnableWithMessageHistory(model,get_session_history)
        
```

```python
config= {"configurable":{"session_id":"chat1"}}
```

```python
response=with_message_history.invoke(
    [HumanMessage(content="Whats my name")],
    config=config
)
```

```python
response.content
```

```python
config1= {"configurable":{"session_id":"chat1"}}
response=with_message_history.invoke(
    [HumanMessage(content="Whats my name")],
    config=config1
)
response.content
```

```python
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder 
prompt=ChatPromptTemplate.from_messages(
    [
        ("system","You are a helful assistant. Answer all the question to the nest of your abiliti",),
        MessagesPlaceholder(variable_name="messages")
        
    ]
)
chain= prompt | model 
```

```python
chain.invoke(
    {
        "messages":[
            HumanMessage(content="I am niraj and i am a cheif software engineer"),
            # AIMessage(content="Your name is Niraj."),
            # HumanMessage(content="What do I do?")
        ]
    }
)
```

```python
with_message_history= RunnableWithMessageHistory(chain, get_session_history)
config= {"configurable":{"session_id":"chat3"}}
response= with_message_history.invoke(
    [HumanMessage(content="Whats my name")],
    config=config
)
```

```python
response.content
```
